```python
import pandas as pd
import numpy as np

import datetime

from scipy.optimize import nnls as nnls
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, mean_absolute_percentage_error
```

```python
file_name = r".\Overall View\All Data.xlsx"

raw_frame = pd.read_excel(file_name, index_col = "Timestamp")

# raw_frame[["Total HVAC", "Total UPS"]].plot()
raw_frame["Const"] = 1
sites = raw_frame["Site"].unique()
sites, raw_frame.columns
```

```python
base_start = "2021-10-01"
base_end = "2021-12-30"

post_start = "2022-01-01"
post_end = "2023-03-01"

Xs = ["Total UPS", "Const", "CDD 18"]
```

```python
file_name = r".\Data\Weather\JB 18 CDD.csv"
JB_weather = pd.read_csv(file_name, skiprows = 6, parse_dates = True, index_col = "Week starting")
JB_weather.index = JB_weather.index + datetime.timedelta(days = 7)


file_name = r".\Data\Weather\CT 18 CDD.csv"
CT_weather = pd.read_csv(file_name, skiprows = 6, parse_dates = True, index_col = "Week starting")
CT_weather.index = CT_weather.index + datetime.timedelta(days = 7)

file_name = r".\Data\Weather\DB 18 CDD.csv"
DB_weather = pd.read_csv(file_name, skiprows = 6, parse_dates = True, index_col = "Week starting")
DB_weather.index = DB_weather.index + datetime.timedelta(days = 7)
```

```python
temp_frame_dict = {'CT1':CT_weather,
                   'CT2':CT_weather,
                   'DB1':DB_weather,
                   'JB1 East':JB_weather,
                   'JB1 West':JB_weather,
                   'JB2':JB_weather,
                   'JB3':JB_weather,
                   'Total':JB_weather}
```

```python
costs = []
savings = []
totals = []
r2s = []

for site in sites:
    
    
    print ("Site {0}".format(site))
    frame = raw_frame[raw_frame["Site"] == site].copy()
    
    temp_frame = temp_frame_dict[site]
    
    frame = pd.concat([frame, temp_frame], axis = 1)
    
    frame[["Total HVAC", "Total UPS"]]["2021":].plot(title = site)
    base_frame = frame[frame["Site"] == site][base_start:base_end].copy()
    base_frame[["Total HVAC", "Total UPS"]]["2021":].plot(title = site)
    
    
    
    X = base_frame[Xs]
    Y = base_frame["Total HVAC"]

    # X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3)
    X_train = X
    Y_train = Y
#     print( 3 * X_train.shape[1])
    results = nnls(X_train, Y_train, maxiter = 20)

    print ("Coeffs and Residual: ", results)

    base_frame["Pred"] = 0
    frame["Pred"] = 0
    for factor, coeff in zip(Xs, results[0]):
        base_frame["Pred"] += coeff * base_frame[factor]
        frame["Pred"] += coeff * frame[factor]


    r2 = r2_score(base_frame["Total HVAC"], base_frame["Pred"])
    print ("R2: ", r2)
    if site != "Total":
        r2s.append(r2)
    
    base_frame[["Total HVAC", "Pred", "Total UPS"]].plot(title = site)

#     base_end



    frame[["Total HVAC", "Pred", "Total UPS"]][post_start:post_end].plot(title = site) #"2022-03-01"

    post_baseline = frame[post_start:post_end].copy()
    post_baseline["Diff"] = post_baseline["Total HVAC"] - post_baseline["Pred"]
    post_baseline["CUSUM"] = post_baseline["Diff"].cumsum(axis = 0)

    post_baseline["CUSUM"][-1]

    cumsum_dates = np.array(post_baseline.index)
    cumsum_vals = post_baseline["CUSUM"].values
    cumsum_vals = np.insert(cumsum_vals, 0, 0)
    cumsum_dates = np.insert(cumsum_dates, 0, "2021-12-05")
    # cumsum_vals, cumsum_dates
    plt.plot(cumsum_dates, cumsum_vals)
    plt.title(site + " CUSUM")

    Power_saving = -post_baseline["CUSUM"][-1]*24*7
    print( "Approximately {} kWh Saved".format(round(Power_saving,0)))
    if site != "Total":
        savings.append(Power_saving)

    est_power_saving_cost = Power_saving * 1.2
    print( "Approximately R{} Saved".format(round(est_power_saving_cost,2)))
    if site != "Total":
        costs.append(est_power_saving_cost)

    all_used = post_baseline["Total Incoming"].sum()*24*7
    print( "Approximately {} kWh Used Total".format(round(all_used,0)))
    if site != "Total":
        totals.append(all_used)

    power_perc = Power_saving/(all_used)*100
    print("Approximately {}% power saved".format(power_perc))
    
    print()
    
    storing_frame = frame["2021":].copy()
    storing_frame.drop("Predicted HVAC", axis = 1, inplace = True)
    storing_frame.drop("Pred", axis = 1, inplace = True)
    storing_frame["Total UPS"] = storing_frame["Total UPS"] * 24 * 7 
    storing_frame["Total HVAC"] = storing_frame["Total HVAC"] * 24 * 7 
    storing_frame["Total Incoming"] = storing_frame["Total Incoming"] * 24 * 7 
    storing_frame.to_excel(r".\Data\Weather\Combined {0}.xlsx".format(site))
```

```python
total_saving = sum(savings)
total_costs = sum(costs)
total_totals = sum(totals)
```

```python
print("""
With a baseline period of {4} to {5}

kWh Saved = {0}kWh
R saved = R{1}
Total used = {2}kwh
Total Percent saved = {3}%
""".format(total_saving, 
                               total_costs, 
                               total_totals, 
                               total_saving/total_totals*100, base_start, base_end))

for r2, site in zip(r2s, sites):
    print ("R2 for Site {0} = {1}".format(site, r2))
```
