```python
column_list = []
tems_list = []
paired = []
for column in data.columns:
    col = column.split("#")[0].split('.')[-1]
    tem = column.split("#")[0].split('.')[0].split('_')
    
    column_list.append(col)
#     print (tem)
    for sub in tem:
        if "TEM" in sub:
            tem_found = sub
            tem = tem_found
    
    tems_list.append(tem)
    
    paired.append([col, column, tem])
# column_list
```

```python
storing_frame = pd.DataFrame(data = paired[1:], columns = ["Name", "Full Name", "TEM"])
storing_frame["Type"] = np.zeros(len(data.columns)-1)
storing_frame["Energy Flow"] = np.zeros(len(data.columns)-1)
```

```python
for row in storing_frame.index:
#     print (row)
    for category in ["UPS", "HVAC", "Generator", "Solar"]:
        if category in storing_frame.loc[row]["Name"]:
            storing_frame.loc[row , "Type"] = category
            

    if "GCL" in storing_frame.loc[row]["Name"]:
        storing_frame.loc[row , "Type"] = "UPS"
    
    if "CCL" in storing_frame.loc[row]["Name"]:
        storing_frame.loc[row , "Type"] = "HVAC"
            
    if "Gen" in storing_frame.loc[row]["Name"]:
        storing_frame.loc[row , "Type"] = "Generator"
    storing_frame["Type"].replace(0,"Mains", inplace = True)
    
    if storing_frame.loc[row,"Type"] in ["Generator", "Mains", "Solar"]:
        storing_frame.loc[row,"Energy Flow"] = "Incoming"
        
    else:
        storing_frame.loc[row,"Energy Flow"] = "Outgoing"
```
