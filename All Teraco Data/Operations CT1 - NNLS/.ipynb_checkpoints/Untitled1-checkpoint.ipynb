```python
import pandas as pd
from scipy.optimize import nnls as nnls
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, mean_absolute_percentage_error
```

```python
frame = pd.read_csv(r".\Data\Master.csv", index_col = "Timestamp", parse_dates = True).drop("Unnamed: 0", axis = 1).resample('d').max().diff()
names = pd.read_csv("Name_list.csv")
```

```python
frame["Const"] = 1
period = 'd'
type_name = "HVAC"
start = "2021-12-01"
end = "2022-01-01"
Xs = ["Total Mains", "Total UPS", "Const", "CDD 16"]
```

```python
for type_name in names["Type"].unique():
    frame["Total {0}".format(type_name)] = frame[names[names["Type"]==type_name]["Full Name"]].sum(axis = 1)
#     frame[names[names["Type"]==type_name]["Full Name"]].plot(marker = '.')

frame["Total Other"] = frame["Total Mains"] + frame["Total Generator"] - frame["Total HVAC"] - frame["Total UPS"]

# print(frame.index)

base_frame = frame.copy()

t_frame = pd.read_csv(r"FACT_CDD_16C+-3C.csv", skiprows = 6, parse_dates = True, index_col = "Date")

new_frame = pd.concat([frame.resample(period).mean(), t_frame.resample(period).mean()], axis = 1).fillna(0)

# print(new_frame.index)
```

```python
frame = frame[:"2022-03-06"]
new_frame = new_frame[:"2022-03-06"]
```

```python
X = new_frame[Xs][start:end]
Y = new_frame["Total HVAC"][start:end]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 1/3)

results = nnls(X_train, Y_train)

X_test["Pred"] = 0
X_train["Pred"] = 0
new_frame["Pred"] = 0

for factor, coeff in zip(Xs, results[0]):
    X_test["Pred"] += coeff * X_test[factor].values
    X_train["Pred"] += coeff * X_train[factor].values
    new_frame["Pred"] += coeff * new_frame[factor].values
    fig = plt.figure()
    new_frame[factor][start:].plot(title = factor)
    print(factor, coeff)
    
fig = plt.figure()
new_frame[["Pred", "Total HVAC"]][start:].plot()
    
Y_test.sort_index(inplace = True)
Y_train.sort_index(inplace = True)
X_test.sort_index(inplace = True)
X_train.sort_index(inplace = True)
    
test_r2 = r2_score(Y_test, X_test["Pred"])
train_r2 = r2_score(Y_train, X_train["Pred"])
total_r2 = r2_score(new_frame["Total HVAC"], new_frame["Pred"])
print("Test R2: {0} \nTraining R2: {1} \nTotal R2: {2}".format(test_r2, train_r2, total_r2))
print(mean_absolute_percentage_error(new_frame["Total HVAC"][start:end], new_frame["Pred"][start:end]))
```
